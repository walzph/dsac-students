```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

## 1. Plot the line fitting example

```python
lf_losses = np.genfromtxt('line_fitting/results.csv', delimiter=' ')
```

```python
# Get data
x = lf_losses[:,0]
y1 = lf_losses[:,2]
y2 = lf_losses[:,3]

# Apply running mean
y1[1:-1] = np.convolve(y1, np.ones((3,))/3, mode='valid')
y2[1:-1] = np.convolve(y2, np.ones((3,))/3, mode='valid')

# Plot training losses
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('iterations')
ax.set_ylabel('loss')
ax.plot(x, y1, label="DSAC winning loss")
ax.plot(x, y2, label="direct prediction loss")
ax.legend()
plt.show()
```
